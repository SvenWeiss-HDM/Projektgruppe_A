```python
import pandas as pd
import numpy as np
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("../data/raw/SpeedDating.csv",delimiter=",")
df.head()
```

```python
df.drop(["iid","id","pid","idg"],axis=1,inplace=True)
```

```python
sns.set(rc={'figure.figsize':(40,80)})
sns.boxplot(df,orient="h")
```

```python
sns.set(rc={'figure.figsize':(20,10)})
g=sns.displot(
    data=df.isna().melt(value_name="NaN"),
    y="variable",
    hue="NaN",
    multiple="fill",
)
g.figure.set_figwidth(20)
g.figure.set_figheight(80) 
```
